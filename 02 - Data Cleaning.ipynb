```python
import pandas as pd
import re
```

```python
starting_base = pd.read_csv('C:/Users/nehad/Downloads/tweets_oct21_base.csv')
len(starting_base)
```

```python
starting_base = starting_base[starting_base['lang'] == 'en']
starting_base = starting_base[['date', 'content']]
len(starting_base)
```

```python
pd.set_option("display.max_colwidth", None)
starting_base.head()
```

```python
pd.set_option("display.max_colwidth", None)
starting_base = starting_base.applymap(lambda x: re.sub(r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*', '', x))
```

```python
import re

def remove_emojis(data):

   emoj = re.compile("["

                      u"\U0001F600-\U0001F64F"  # emoticons
                      u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                      u"\U0001F680-\U0001F6FF"  # transport & map symbols
                      u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                      u"\U00002500-\U00002BEF"  # chinese char
                      u"\U00002702-\U000027B0"
                      u"\U00002702-\U000027B0"
                      u"\U000024C2-\U0001F251"
                      u"\U0001f926-\U0001f937"
                      u"\U00010000-\U0010ffff"
                      u"\u2640-\u2642" 
                      u"\u2600-\u2B55"
                      u"\u200d"
                      u"\u23cf"
                      u"\u23e9"
                      u"\u231a"
                      u"\ufe0f"  # dingbats
                      u"\u3030" 

                       "]+", flags=re.UNICODE)

   return re.sub(emoj, '', data)
```

```python
starting_base['content'] = starting_base['content'].map(remove_emojis)
```

```python
# starting_base = starting_base.replace("\\n", "").replace("\n", " ").replace("\\", "") 
starting_base = starting_base.replace('\\n',' ', regex=True)
starting_base
```

```python
def rm_name(tweet):
    return re.sub('@[\w]+','',tweet)
```

```python
starting_base['content'] = starting_base['content'].map(rm_name)
starting_base
```

```python
def rm_newline(tweet):
    return re.sub('\n','',tweet)


starting_base['content'] = starting_base['content'].map(rm_newline)
starting_base
```

```python
def get_date(date):
    return date.split()[0]
    
starting_base['date'] = starting_base['date'].map(get_date)
starting_base
```

```python
# def rm_hashtag(tweet):
#     return re.sub('#[\w]+','',tweet)

def rm_hash(tweet):
    return re.sub('#','',tweet)



starting_base['content'] = starting_base['content'].map(rm_hash)
starting_base
```

```python
starting_base = starting_base.drop_duplicates()
len(starting_base)
```

```python
#checkpoint 1 - cleaned data
starting_base.to_csv('cleaned_base_oct21.csv', sep=',', index=False)
```
